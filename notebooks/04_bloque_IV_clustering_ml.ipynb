## Carga y escalado

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

DATA_DIR = Path("../datasets")
pd.set_option("display.max_columns", 50)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

df = pd.read_csv(DATA_DIR / "clientes_clustering.csv")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
df.head()
```

## K-Means y silhouette

```python
scores = []
for k in range(2, 8):
    km = KMeans(n_clusters=k, n_init="auto", random_state=42)
    labels = km.fit_predict(X_scaled)
    scores.append({"k": k, "silhouette": silhouette_score(X_scaled, labels)})
scores_df = pd.DataFrame(scores)
display(scores_df)

scores_df.plot(x="k", y="silhouette", marker="o", title="Selección de k por silhouette")
plt.show()
```

## Perfilado de clusters

```python
k_final = int(scores_df.sort_values("silhouette", ascending=False).iloc[0]["k"])
km = KMeans(n_clusters=k_final, n_init="auto", random_state=42)
df_clustered = df.copy()
df_clustered["cluster"] = km.fit_predict(X_scaled)
display(df_clustered.groupby("cluster").mean().round(2))
```

## PCA para visualización

```python
pca = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(X_scaled)
plt.scatter(coords[:,0], coords[:,1], c=df_clustered["cluster"])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters visualizados con PCA")
plt.show()
```

```python

# -------------------------
# ELECCIÓN DE K (MODIFICADO)
# ahora probamos más granularidad
# -------------------------
scores = []

for k in range(2, 11):  # ampliamos rango de clusters
    km = KMeans(
        n_clusters=k,
        n_init=20,        # más inicializaciones para estabilidad
        max_iter=500,     # más iteraciones para mejor convergencia
        random_state=42
    )
    
    labels = km.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, labels)
    
    scores.append({"k": k, "silhouette": sil})

scores_df = pd.DataFrame(scores)
display(scores_df)

scores_df.plot(x="k", y="silhouette", marker="o", title="Selección de k por silhouette")
plt.show()

# -------------------------
# MODELO FINAL
# -------------------------
k_final = int(scores_df.sort_values("silhouette", ascending=False).iloc[0]["k"])

km_final = KMeans(
    n_clusters=k_final,
    n_init=20,
    max_iter=500,
    random_state=42
)

df_clustered = df.copy()
df_clustered["cluster"] = km_final.fit_predict(X_scaled)

# -------------------------
# PERFILADO
# -------------------------
display(df_clustered.groupby("cluster").mean().round(2))

# -------------------------
# VISUALIZACIÓN PCA
# -------------------------
pca = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(X_scaled)

plt.scatter(coords[:, 0], coords[:, 1], c=df_clustered["cluster"])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters visualizados con PCA")
plt.show()


print("Interpretación:\n" \
"He aplicado K-Means para agrupar clientes tras escalar los datos. He probado varios valores de k y elegido el mejor con silhouette.\n"
"El modelo separa bastante bien a los clientes en grupos con perfiles distintos.\n" \
"Con PCA se aprecia la separación, aunque hay algo de solapamiento normal al reducir dimensiones.")
```
